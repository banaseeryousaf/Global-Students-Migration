# DATA PRE-PROCESSING

**Null Value Analysis**

```python
df.isnull().sum()
```

**Handling missing Data**

*Null values in the columns "placement_country" and "placement_company" are filling using 'ffill' and 'bfill' to ensure no null value is left*

```python
df[['placement_country','placement_company']] = df[['placement_country','placement_company']].fillna(method='ffill').fillna(method='bfill')
```

```python
df.isnull().sum()
```

*Imputing Nulls with 'IELTS' in the column "language_proficiency_test"*

```python
df['language_proficiency_test']=df['language_proficiency_test'].fillna('IELTS')
```

```python
df.isnull().sum()
```

All null values are cleared

**Checking Duplicates**

```python
df.duplicated().any()
```

No duplicates found

**Deriving new column "Course_Duration"**

```python
df['course_duration'] = df['graduation_year'] - df['year_of_enrollment']
df
```

**Creating new column 'performance_category'**

```python
bins = [0, 1, 2, 3, 4,5]   
labels = ['poor', 'average', ' Good', 'very good','excellent']

df['performance_category'] = pd.cut(df['gpa_or_score'], bins=bins, labels=labels, right=False)
print(df[['gpa_or_score', 'performance_category']])
```

```python
df['placement_status_binary']=df['placement_status'].apply(lambda x: 1 if x=='Placed' else 0)
```

# Key Insights

**1. Global Migration Hubs and Corridors**   
The data reveals a shift from traditional Western education hubs toward emerging global centers.  
- The UAE as a Global Magnet: The UAE is the most popular destination country in this dataset. It acts as a primary hub for students from diverse        backgrounds, specifically attracting large flows from Germany (72), Russia (71), and the UK (66).  
- Russia as a Major Origin: Russia is the top country of origin for emigrating students, with significant numbers moving to the UAE, Ireland (66),       and Germany (67).  
- Established Corridors: Traditional routes remain very strong, particularly the India to UK corridor (72 students), reflecting long-standing            educational and professional ties.  
- India → UK and Germany → UAE are the strongest migration routes.
